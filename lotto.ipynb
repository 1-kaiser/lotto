```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load data
```

```python
df = pd.read_csv('lotto_games.csv')
```

```python
## Data Cleaning & Preprocessing
```

```python
''' 
Note: 
We need to drop unused columns: no_draw, no_draw_reason, created_at, updated_at
We need to convert drawdate into datetime and drawtime into time
'''
```

```python
## Drop useless columns
```

```python
data_clean = df.drop(columns=['no_draw', 'no_draw_reason', 'created_at', 'updated_at'])
data_clean
```

```python
# Convert date and time
```

```python
data_clean['draw_date'] = pd.to_datetime(data_clean['draw_date'])
data_clean['draw_time'] = pd.to_datetime(data_clean['draw_time'], format="%H:%M:%S").dt.time
```

```python
## Extract date parts
```

```python
data_clean['year'] = data_clean['draw_date'].dt.year
data_clean['month'] = data_clean['draw_date'].dt.month
data_clean['day'] = data_clean['draw_date'].dt.day
data_clean['weekday'] = data_clean['draw_date'].dt.day_name()
```

```python
## Count numbers in each draw
```

```python
data_clean['num_count'] = data_clean['winning_numbers'].apply(lambda x: len(x.split('-')))
```

```python
## EDA
```

```python
## Game distribution
```

```python
plt.figure(figsize=(12, 6))
data_clean["formatted_game_with_time"].value_counts().head(15).plot(kind="bar")
plt.title("Top 15 Lotto Games by Draw Count")
plt.xlabel('Games')
plt.ylabel("Number of Draws")
plt.show()
```

```python
## Jackpot winners distribution
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data_clean['jackpot_winners'], bins=50)
plt.title('Distribution of Jackpot Winners')
plt.xlabel('Number of winners')
plt.ylabel('Frequency')
plt.xlim(0, 100)
plt.show()
```

```python
## Jackpot trend over time
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(data=data_clean, x="draw_date", y="jackpot_amount_in_php", hue="formatted_game_with_time", legend=False)
plt.ylim(0, 5e8)
plt.title("Jackpot Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Jackpot Amount")
plt.show()
```
